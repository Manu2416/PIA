1.	Dado un fichero ‘frutas.txt’ con una palabra por línea, crear un fichero de salida con las frutas ordenadas.

```python


f = open("./docsEj/frutas.txt","r",encoding="utf-8")
frutas = f.readlines()

frutas.sort()
f.close()

f2 = open("./docsEj/frutasordenadas.txt","w")
for fruta in frutas:
    f2.write(fruta)
f2.close()


    
```

2.	Copiar un fichero de datos de personas. Crea una copia exacta de 'personas.txt' en 'personas_copia.txt' recorriendo las líneas y escribiéndolas una a una. 

```python
with open("docsEj/personas.txt","r",encoding="utf-8") as f:
    personas = f.readlines()
    f.close()
    copia = open("docsEj/copia_personas.txt","w")
    copia.writelines(personas)   
```

3.	Abre el archivo ‘personas.txt’ y añade datos de nuevas personas (nombre y edad) al final. Hazlo preguntando al usuario hasta que este diga que no quiere seguir.

```python
with open("docsEj/personas.txt","a") as f:
    while True:
        seguir = input("quieres seguir?")
        if seguir == "si":
            nombre = input("dame un nombre")
            edad = input("dame una edad")
            escribir= (nombre + "," + edad )
         
            f.write(escribir + "\n")
            f.close
        else:
            break


    
```

4.	Comprobar si un archivo llamado ‘alumnos.txt’ existe y si no, crearlo. Pide al usuario Nombre de alumno, módulo y nota para guardar en el fichero una línea por alumno, con los datos separados por comas.

```python
import os
path = "docsEj/personas.txt"
if os.path.exists(path) :
    print("ya existe")
else:
    f = open(path,"a")
    nombre = input("dame un nombre")
    modulo = input("dame un modulo")
    nota = input("nota")
    escribir= (nombre + "," + modulo + "," + nota )
    
    f.write(escribir + "\n")
    f.close()
  

```

5.	Abre el fichero anterior, que contiene notas de los alumnos y recórrelo para calcular la media de las notas de cada módulo.

```python

listas = []

with open("docsEj/personas.txt", "r", encoding="utf-8") as f:
    for row in f:
        listas.append(row) 

datos = listas[1:]

diccionario = {}

for fila in datos:
    partes = fila.split(",")  
    asignatura = partes[1]
    nota = float(partes[2]) 

    if asignatura not in diccionario:
        diccionario[asignatura] = {"suma": 0, "contador": 0}

    diccionario[asignatura]["suma"] += nota
    diccionario[asignatura]["contador"] += 1


for asignatura, valores in diccionario.items():
    media = valores["suma"] / valores["contador"]
    print(f"Media de {asignatura} es {media}")

```
