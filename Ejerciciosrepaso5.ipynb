Después pregunta al usuario de qué material quiere conocer el stock y muéstralo por pantalla.

```python
import csv

listas = []


with open("./docsEj/stock.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        listas.append(row)
    #Aqui creo dos listas una con el nombre y otras con los datos
    nombre_materiales = listas[0]
    datos = listas[1:]
    #inicializo una lista vacia para contar el stock total y un diccionario para poder almacenarlo
    stock =[0,0,0]
    diccionario = {}
    #cuento cada stock de material y lo sumo
    for dato in datos:
        for i in range(len(dato)):
            stock[i] += int(dato[i])
    #por cada material que tengo le asigno su stock en el diccionario
    for material in nombre_materiales:
        for i in range(len(stock)):
            diccionario[material]= stock[i]
    print(diccionario)
    #Creo una excepcecion si el nombre d ela key no existe en mi diccionario saltara
    try:
        respuesta = input("que stock quieres saber?")
        print (f"el stock de {respuesta} es {diccionario[respuesta]}")
    except KeyError:
        print("No existe ese material en el stock")
```
