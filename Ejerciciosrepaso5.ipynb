1.	Haciendo uso del fichero stock.csv realiza un programa que abra el fichero, lea sus datos y los almacene en una lista de diccionarios. Después imprime en pantalla el stock de cada material. Prueba a hacerlo con y sin usar la estructura “with” de acceso a fichero.

```python
import csv

listas = []

# Leer el CSV completo
with open("./docsEj/stock.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        listas.append(row)

# La primera fila son los encabezados (nombres de productos)
encabezados = listas[0]

# Las demás filas son los datos numéricos
datos = listas[1:]

# Creamos un diccionario para acumular los totales
diccionario = {}

for nombre in encabezados:
    diccionario[nombre] = 0

# Recorremos todas las filas de datos y sumamos por columna
for fila in datos:
    for i, valor in enumerate(fila):
        diccionario[encabezados[i]] += int(valor)

# Mostramos los resultados
print("Totales de cada producto:")
for producto, total in diccionario.items():
    print(f"{producto}: {total}")

  
```
