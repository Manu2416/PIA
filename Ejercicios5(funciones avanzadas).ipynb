1.       Calculadora Básica: Crea funciones individuales para sumar, restar, multiplicar y dividir dos números.

La función principal calculadora deberá recibir tres argumentos: dos números y una operación, y en base a la operación, llamará a la función correspondiente.

```python
def sumar(num1,num2):
    return num1+num2

def restar(num1,num2):
    return num1-num2

def multiplicar(num1,num2):
    return num1*num2

def dividir(num1,num2):
    return num1/num2

def menu():
        print("calculadora")
        print("1. sumar")
        print("2. restar")
        print("3. multiplicar")
        print("4 dividir")
        opcion = int(input("Que quieres hacer?"))
        num1 = int(input("dame un numero"))
        num2 = int(input("dame otr numero"))
        if opcion == 1:
            print(sumar(num1,num2))
        elif opcion == 2:
            print(restar(num1,num2))
        elif opcion == 3:
            print(multiplicar(num1,num2))
        elif opcion == 4:
            print(dividir(num1,num2))
        else:
            print ("caractero no valido")


menu()
```

2.       Análisis de Texto: Escribe una función que cuente el número de palabras en un texto.

Otra función que cuente la cantidad de veces que aparece una palabra específica.

La función principal analizar_texto recibirá un texto y una palabra a buscar, y usará las funciones anteriores para imprimir el total de palabras y la frecuencia de la palabra buscada.

```python
def n_palabras(frase):
        palabras = frase.split(" ")
        n=0
        for palabra in palabras:
               n += 1

        return n


def n_apariciones(frase,palabra_especificar):
        palabras = frase.split(" ")
        n=0
        for palabra in palabras:
               if palabra == palabra_especificar:
                n += 1

        return n
def analizar_texto(texto,palabra):
      print("las palarbas que hay son",n_palabras(texto))
      print("la palabra que buscas aparece ",n_apariciones(texto,palabra))


analizar_texto("hola que que tal","que")
```

3.      Sistema de Calificaciones: Realiza una función para calcular el promedio de una lista de calificaciones.

Otra función que determine la calificación más alta.

La función principal informe_calificaciones tomará una lista de calificaciones, y usando las funciones anteriores, imprimirá el promedio y la calificación más alta.

```python
def media (notas):
    valor = 0
    n_notas = 0
    for nota in notas:
        n_notas += 1
        valor += nota

    media = valor / n_notas

    return media


def notas_alta(notas):
    nota_mayor = notas[0]
    for nota in notas:
        if nota_mayor <= nota:
            nota_mayor = nota
    return nota_mayor

notas = [1,4,6,7,8,4,3]
def informe(notas):
    print("la media es ", media(notas))
    print ("la mas alta es ", notas_alta(notas))

informe(notas)
```

4.      Conversor de Unidades: Crea funciones individuales para convertir kilómetros a millas, kilogramos a libras, y litros a galones.

La función principal conversor_unidades pide al usuario el tipo de conversión y el valor, y utiliza la función adecuada para realizar la conversión e imprimir el resultado.

```python
# Funciones de conversión
def km_a_millas(km):
    return km * 0.621371

def kg_a_libras(kg):
    return kg * 2.20462

def litros_a_galones(litros):
    return litros * 0.264172

# Función principal
def conversor_unidades():
    print("=== Conversor de Unidades ===")
    print("1. Kilómetros a millas")
    print("2. Kilogramos a libras")
    print("3. Litros a galones")

    opcion = int(input("Elige una opción (1-3): "))
    valor = float(input("Introduce el valor a convertir: "))

    if opcion == 1:
        resultado = km_a_millas(valor)
        print(f"{valor} km son {resultado:.2f} millas.")
    elif opcion == 2:
        resultado = kg_a_libras(valor)
        print(f"{valor} kg son {resultado:.2f} libras.")
    elif opcion == 3:
        resultado = litros_a_galones(valor)
        print(f"{valor} litros son {resultado:.2f} galones.")
    else:
        print("Opción no válida.")

# Llamada al programa
conversor_unidades()
```
